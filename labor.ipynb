```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
########PROBLEMA 1#####

def activar_canales_color(imagen, color):
    # Verificamos que la imagen esté en el formato adecuado
    if imagen.shape[2] != 3:
        raise ValueError("La imagen debe tener 3 canales de color (RGB).")
    
    # Creamos una copia de la imagen original
    imagen_resultado = imagen.copy()
    
    # Aplicamos las máscaras según el valor de 'color'
    if color == 1:
        # Solo azul
        imagen_resultado[:, :, 1] = 0  # Canal verde
        imagen_resultado[:, :, 2] = 0  # Canal rojo
    elif color == 2:
        # Solo verde
        imagen_resultado[:, :, 0] = 0  # Canal azul
        imagen_resultado[:, :, 2] = 0  # Canal rojo
    elif color == 3:
        # Solo rojo
        imagen_resultado[:, :, 0] = 0  # Canal azul
        imagen_resultado[:, :, 1] = 0  # Canal verde
    elif color == 10:
        # Rojo y verde
        imagen_resultado[:, :, 0] = 0  # Canal azul
    elif color == 20:
        # Verde y azul
        imagen_resultado[:, :, 2] = 0  # Canal rojo
    elif color == 30:
        # Azul y rojo
        imagen_resultado[:, :, 1] = 0  # Canal verde
    else:
        raise ValueError("El valor del parámetro 'color' no es válido. Debe ser 1, 2, 3, 10, 20, o 30.")
    
    return imagen_resultado

# Cargar la imagen de ejemplo
img = os.path.abspath('imagenes\ejer1.jpg')
imagen = cv2.imread(img)

# Convertir la imagen de BGR (por defecto en OpenCV) a RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Aplicar la función con un valor de ejemplo para 'color'
imagen_resultado = activar_canales_color(imagen_rgb, 20)

# Mostrar la imagen original y la imagen resultado
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagen con Color Activado')
plt.imshow(imagen_resultado)
plt.axis('off')

plt.show()
```

```python
########PROBLEMA 2#####

def combinar_imagenes_a_color(imagen1, imagen2, imagen3):
    # Verificar que las imágenes tengan las mismas dimensiones
    if imagen1.shape != imagen2.shape or imagen2.shape != imagen3.shape:
        raise ValueError("Las imágenes deben tener las mismas dimensiones.")
    
    # Combinar las imágenes en una sola imagen a color
    imagen_color = np.stack((imagen1, imagen2, imagen3), axis=-1)
    
    return imagen_color

# Cargar las imágenes en escala de grises
imagen1 = cv2.imread(os.path.abspath('imagenes\perro\perro_salida_gray_rojo.jpg'), cv2.IMREAD_GRAYSCALE)
imagen2 = cv2.imread(os.path.abspath('imagenes\perro\perro_salida_gray_verde.jpg'), cv2.IMREAD_GRAYSCALE)
imagen3 = cv2.imread(os.path.abspath('imagenes\perro\perro_salida_gray_azul.jpg'), cv2.IMREAD_GRAYSCALE)

# Combinar las imágenes en una imagen a color
imagen_color = combinar_imagenes_a_color(imagen1, imagen2, imagen3)

# Mostrar la imagen resultado
plt.imshow(imagen_color)
plt.title('Imagen Combinada a Color')
plt.axis('off')
plt.show()
```

```python
#########PROBLEMA 3#####
def crear_escala_grises_3d(imagen):
    # Verificar que la imagen tenga 3 canales (RGB)
    if imagen.shape[2] != 3:
        raise ValueError("La imagen debe tener 3 canales de color (RGB).")
    
    # Separar los canales de la imagen
    canal_azul = imagen[:, :, 0]
    canal_verde = imagen[:, :, 1]
    canal_rojo = imagen[:, :, 2]
    
    return canal_azul, canal_verde, canal_rojo

# Cargar la imagen
imagen = cv2.imread(os.path.abspath('imagenes\ejer2.jpg'))
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Crear la escala de grises en 3D
canal_azul, canal_verde, canal_rojo = crear_escala_grises_3d(imagen_rgb)

# Mostrar los canales resultantes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(canal_azul, cmap='gray')
axs[0].set_title('Canal Azul')
axs[0].axis('off')

axs[1].imshow(canal_verde, cmap='gray')
axs[1].set_title('Canal Verde')
axs[1].axis('off')

axs[2].imshow(canal_rojo, cmap='gray')
axs[2].set_title('Canal Rojo')
axs[2].axis('off')

plt.show()
```

```python

#########PROBLEMA 4#####

def mostrar_histogramas(imagen):
    # Convertir la imagen a escala de grises
    imagen_gris = np.mean(imagen, axis=2).astype(np.uint8)
    
    # Calcular los histogramas
    hist_azul = np.histogram(imagen[:, :, 0], bins=256, range=(0, 256))[0]
    hist_verde = np.histogram(imagen[:, :, 1], bins=256, range=(0, 256))[0]
    hist_rojo = np.histogram(imagen[:, :, 2], bins=256, range=(0, 256))[0]
    hist_gris = np.histogram(imagen_gris, bins=256, range=(0, 256))[0]
    
    # Calcular las medias
    media_azul = np.mean(imagen[:, :, 0])
    media_verde = np.mean(imagen[:, :, 1])
    media_rojo = np.mean(imagen[:, :, 2])
    media_gris = np.mean(imagen_gris)
    
    # Mostrar los histogramas
    fig, axs = plt.subplots(4, 1, figsize=(10, 20))
    
    axs[0].bar(range(256), hist_azul, color='blue', alpha=0.6)
    axs[0].axvline(media_azul, color='red', linestyle='dashed', linewidth=1)
    axs[0].set_title('Histograma Canal Azul')
    
    axs[1].bar(range(256), hist_verde, color='green', alpha=0.6)
    axs[1].axvline(media_verde, color='red', linestyle='dashed', linewidth=1)
    axs[1].set_title('Histograma Canal Verde')
    
    axs[2].bar(range(256), hist_rojo, color='red', alpha=0.6)
    axs[2].axvline(media_rojo, color='blue', linestyle='dashed', linewidth=1)
    axs[2].set_title('Histograma Canal Rojo')
    
    axs[3].bar(range(256), hist_gris, color='gray', alpha=0.6)
    axs[3].axvline(media_gris, color='red', linestyle='dashed', linewidth=1)
    axs[3].set_title('Histograma Escala de Grises')
    
    plt.show()

# Cargar la imagen
imagen = cv2.imread(os.path.abspath('imagenes\ejer3.jpg'))
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Mostrar los histogramas
mostrar_histogramas(imagen_rgb)
```

#########PROBLEMA 5#####

La escala de grises ponderada utiliza una combinación de los canales RGB para reflejar mejor la percepción humana del brillo. 
Los pesos comunes son 0.299 para el rojo, 0.587 para el verde y 0.114 para el azul.


```python
def escala_grises_ponderada(imagen):
    # Verificar que la imagen tenga 3 canales (RGB)
    if imagen.shape[2] != 3:
        raise ValueError("La imagen debe tener 3 canales de color (RGB).")
    
    # Aplicar la ponderación
    escala_grises = (0.299 * imagen[:, :, 2] + 0.587 * imagen[:, :, 1] + 0.114 * imagen[:, :, 0]).astype(np.uint8)
    
    return escala_grises

# Cargar la imagen
imagen = cv2.imread(os.path.abspath('imagenes\ejer4.jpg'))
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Convertir a escala de grises ponderada
imagen_gris_ponderada = escala_grises_ponderada(imagen_rgb)

# Mostrar la imagen resultado
plt.imshow(imagen_gris_ponderada, cmap='gray')
plt.title('Imagen en Escala de Grises Ponderada')
plt.axis('off')
plt.show()
```

#########PROBLEMA 6#####

* El espacio de color HSV (Hue, Saturation, Value) representa los colores de una manera que es más intuitiva para los humanos. 
* El matiz (Hue) representa el color, la saturación (Saturation) representa la pureza del color y el valor (Value) representa el brillo del color.
* El matiz es un ángulo en un círculo de color (0 a 360 grados), la saturación varía de 0 a 100% y el valor varía de 0 a 100%.
* Para convertir una imagen de RGB a HSV en OpenCV, se puede usar la función cv2.cvtColor.


```python
# Cargar la imagen
imagen = cv2.imread(os.path.abspath('imagenes\ejer5.jpg'))
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Convertir la imagen de RGB a HSV
imagen_hsv = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2HSV)

# Mostrar la imagen en el espacio de color HSV
plt.imshow(imagen_hsv)
plt.title('Imagen en Espacio de Color HSV')
plt.axis('off')
plt.show()
```
